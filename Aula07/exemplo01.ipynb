# Exemplo 01 - Selecao por Roleta (Roulette Wheel Selection)

Este e um dos metodos de selecao mais classicos em **Algoritmos Geneticos**.

---

## Ideia Central

Cada individuo da populacao recebe uma **"fatia" da roleta** proporcional ao seu valor de aptidao (fitness).

- Quanto **MAIOR** a aptidao → **MAIOR** a fatia → **MAIOR** a probabilidade de ser selecionado.

## Analogia

Imagine uma **roleta de cassino** dividida em fatias de tamanhos diferentes. Giramos a roleta e vemos onde a bolinha para.

- Individuos mais aptos ocupam fatias maiores
- Mesmo os menos aptos ainda tem uma pequena chance!

## Por que isso importa?

- Mantem a **diversidade genetica** (nao descarta totalmente os mais fracos)
- Da vantagem aos melhores **sem ser deterministico**
- E a base para entender outros metodos de selecao mais avancados

```python
import random
```

---
## Passo 1: Definir a populacao e os valores de aptidao (fitness)

Cada individuo e representado por uma letra (A, B, C, D).  
Em um AG real, cada individuo seria um **cromossomo** (ex: vetor binario).

Os valores de **fitness** indicam o quao bom cada individuo e:  
- **B** tem fitness 3 (o melhor)
- **C** tem fitness 0.5 (o pior)

```python
# Populacao: 4 individuos
populacao = ['A', 'B', 'C', 'D']

# Valores de aptidao (fitness) de cada individuo
# Quanto MAIOR o valor, melhor o individuo resolve o problema
fitness = [1, 3, 0.5, 2]

print("Populacao:", populacao)
print("Fitness:  ", fitness)
```

---
## Passo 2: Visualizar as probabilidades de selecao

A probabilidade de cada individuo ser selecionado e:

$$P(i) = \frac{fitness_i}{\sum fitness}$$

Vamos calcular e visualizar:

```python
# Calcular a probabilidade de cada individuo
soma_total = sum(fitness)  # 1 + 3 + 0.5 + 2 = 6.5

print("=" * 50)
print("ROLETA - PROBABILIDADES DE SELECAO")
print("=" * 50)

for individuo, nota in zip(populacao, fitness):
    probabilidade = nota / soma_total * 100
    barra = "|" * int(probabilidade / 2)  # Barra visual proporcional
    print(" ", individuo, ": fitness =", nota, " -> ", round(probabilidade, 1), "%", barra)

print("\n  Soma total de fitness:", soma_total)
print("=" * 50)
```

---
## Passo 3: Implementar a selecao por roleta

O algoritmo funciona assim:

1. **Soma** todos os valores de fitness (tamanho total da roleta)
2. **Gera** um numero aleatorio entre 0 e a soma total ("gira a roleta")
3. **Percorre** os individuos, acumulando seus fitness
4. O individuo cuja soma acumulada **ultrapassa** o ponto sorteado e o selecionado

![Analogia da Roleta](https://upload.wikimedia.org/wikipedia/commons/2/2a/Fitness_proportionate_selection_example.png)

```python
def selecao_roleta(pop, fit):
    """
    Seleciona UM individuo da populacao usando o metodo da roleta.

    Parametros:
        pop (list): Lista de individuos da populacao.
        fit (list): Lista de valores de fitness correspondentes.

    Retorna:
        O individuo selecionado.
    """

    # Passo 3.1: Calcular o tamanho total da roleta
    soma_total = sum(fit)

    # Passo 3.2: "Girar a roleta" - sortear um ponto aleatorio
    # random.uniform(a, b) retorna um float aleatorio entre a e b
    ponto_sorteio = random.uniform(0, soma_total)

    # Passo 3.3: Percorrer a roleta ate encontrar o individuo selecionado
    soma_atual = 0
    for individuo, nota in zip(pop, fit):
        soma_atual += nota  # Acumula o fitness

        # Quando a soma acumulada atinge ou ultrapassa o ponto sorteado,
        # significa que a "bolinha" parou na fatia deste individuo
        if soma_atual >= ponto_sorteio:
            return individuo

print("Funcao selecao_roleta() definida com sucesso!")
```

---
## Passo 4: Testar a selecao (1 vez)

Vamos rodar a selecao **uma unica vez** para ver quem e selecionado.  
**Rode esta celula varias vezes** e observe que o resultado muda!

```python
# Selecionar UM individuo
pai_selecionado = selecao_roleta(populacao, fitness)
print("Individuo selecionado:", pai_selecionado)

# EXPERIMENTE: rode esta celula varias vezes!
# B deve aparecer com mais frequencia (fitness = 3)
```

---
## Passo 5: Verificacao experimental

Se rodarmos a selecao **muitas vezes** (10.000), as frequencias observadas devem se **aproximar** das probabilidades teoricas.

Isso e a **Lei dos Grandes Numeros** em acao!

```python
# Verificacao com 10.000 sorteios
contagem = {ind: 0 for ind in populacao}  # Contador para cada individuo
num_sorteios = 10000

for _ in range(num_sorteios):
    selecionado = selecao_roleta(populacao, fitness)
    contagem[selecionado] += 1

# Exibir resultados experimentais vs teoricos
print("=" * 50)
print("VERIFICACAO: 10.000 sorteios")
print("=" * 50)

for individuo, nota in zip(populacao, fitness):
    prob_teorica = nota / sum(fitness) * 100
    prob_experimental = contagem[individuo] / num_sorteios * 100
    print(" ", individuo, ": teorico =", round(prob_teorica, 1), "%  |  experimental =", round(prob_experimental, 1), "% (", contagem[individuo], "vezes)")

print("=" * 50)
print("\nObserve como os valores experimentais se aproximam dos teoricos!")
```

## Discussao em aula

**Perguntas para reflexao:**

1. O que acontece se um fitness for **zero**? O individuo nunca seria selecionado.
2. E se um fitness for **negativo**? A roleta quebra! (motivacao para normalizacao)
3. **Pressao seletiva**: Se um individuo domina demais, os outros quase nao sao selecionados → convergencia prematura.
4. **Alternativas**: Selecao por torneio, ranking, etc.
